```python
import time
import pandas as pd

```

```python
csv_path = "miami_crimes/Resources/Municipal_Police_Station.csv"
stations_df = pd.read_csv(csv_path)

stations_df.head()
```

```python
stations_df = stations_df.rename(columns = {'FOLIO' : 'Rec_Nbr', 'NAME':'school', 'ADDRESS':'address', 'PHONE':'phone','CHIEF':'principal', 'LAT':'latitude', 'LON':'longitude'})
stations_df.head()
```

```python
for index, row in stations_df.iterrows():
    addr = stations_df.loc[index, 'address']
    city = stations_df.loc[index, 'CITY']
    zip = stations_df.loc[index, 'ZIPCODE']
    full_address = f'{addr} {city} {zip}'
    print(index, full_address)
    stations_df.loc[index, 'address'] = full_address
    
stations_df
```

```python
stations_df['type']='p'
stations_df['lat_dir']=''
stations_df['lng_dir']=''


stations_df
```

```python
pol_stations_df = stations_df[['Rec_Nbr', 'school', 'address', 'phone', 'principal', 'type', 'latitude', 'longitude', 'lat_dir', 'lng_dir']]
 

pol_stations_df
```

```python
csv_path = "output/dadeschools_lat_lng.csv"
schools_df = pd.read_csv(csv_path)
```

```python
combined = schools_df.append(pol_stations_df, ignore_index = True) 
combined
```
