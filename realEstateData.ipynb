```python
import time
import pandas as pd
```

```python
csv_path = "Resources/Summary_AllHomes.csv"
homes_by_zip_df = pd.read_csv(csv_path, sep=',',encoding='ANSI')

homes_by_zip_df.head()
```

```python
miami_homes_by_zip_df = homes_by_zip_df[homes_by_zip_df['County']=='Miami-Dade County']
miami_homes_by_zip_df
```

```python
csv_path_zip = "Resources/zip_code_miami_dade.csv"
zip_code_df = pd.read_csv(csv_path_zip)

zip_code_df
```

```python
combined_df = pd.merge(miami_homes_by_zip_df, zip_code_df, left_on='RegionName', right_on='Zip_Code')
```

```python
combined_df
```

```python
combined_df.to_csv('Resources/dade_homes.csv')
```

```python
csv_path_schools = "Resources/dadeSchoolsStations.csv"
schools_stations_df = pd.read_csv(csv_path_schools)

schools_police_df = schools_stations_df.rename(columns={"latitude":"Latitude", "longitude":"Longitude"})
schools_df = schools_police_df[['Latitude', 'Longitude', 'school','address','phone','principal', 'type']]
schools_df
```

```python
combined_df['type']='r'
combined_df['phone'] =' '
combined_df['address'] =' '
combined_df['principal']=' '
combined_df['school']=' '
combined_df
```

```python
concat_df = pd.concat([combined_df, schools_df], ignore_index=True) 
#pd.concat([s1, s2], ignore_index=True)
concat_df
```

```python
concat_df.fillna(value=' ', inplace = True)
concat_df
```

```python
concat_df.to_csv('Resources/dade_homes_schools_stations.csv')
```
