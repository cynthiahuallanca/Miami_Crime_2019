```python
import pandas as pd
# import matplotlib.pylab as plt
import matplotlib.pyplot as plt
from datetime import datetime
%matplotlib inline
import numpy as np
```

```python
df = pd.read_csv('Resources/jailbook_cleaned.csv')
df.head()
```

```python
# Create a new column for Age categories with ranges: 0-25 , 26-50 , >50 where the value is 1 if it is true and 0 if not
df['Age<25'] = np.where(df['Age']<=25, 1,0)
df['Age26-50'] = np.where((df['Age']>25) & (df['Age']<=50), 1,0)
df['Age>50'] = np.where(df['Age']>50, 1,0)

# df.head()
```

```python
df.tail()
```

```python
# DELETE AFTER. 
# this part will be added in cleaning process. 

df['Booking_year_month'] = df['Booking_year_month'].astype(str)
df['Booking_month'] = df['Booking_year_month'].str.split('-').str.get(1)

df.head()
```

```python
df['Booking_month'] = df['Booking_month'].astype(int)
```

```python
df.dtypes
```

```python
df.head()
```

```python
# Create Seasons columns where: Summer= 7, 8, 9. Spring=4,5,6. Winter=1,2,3. Autum=10,11,12.
df['Summer'] = np.where((df['Booking_month']>=7) & (df['Booking_month']<=9), 1,0)
df['Spring'] = np.where((df['Booking_month']>=4) & (df['Booking_month']<=6), 1,0)
df['Winter'] = np.where((df['Booking_month']>=1) & (df['Booking_month']<=3), 1,0)
df['Autum'] = np.where((df['Booking_month']>=10) & (df['Booking_month']<=12), 1,0)
```

```python
df.tail()
```

```python
# DELETE AFTER. 
# this part will be added in cleaning process. 

# Drop innecesary columns
df.drop(['Unnamed: 0','Address','City','State','Crime_Family2','Crime_Family'], axis=1, inplace=True)
```

```python
df["total_crime"] = df['Age<25'] + df['Age26-50'] + df['Age>50']
df.head()
```

```python
# Filter by State of Florida only (dropping NA also)  
df_assault = df.loc[df['Crime_Family1'] == "ASSAULT & BATTERY",:]
df_assault.head()
```

```python
df_assault_drop = df_assault.copy()
```

```python
df_assault_drop.drop(['Defendant','Zip','Booking_Date','Date_of_birth','Crime_Family1','Booking_year_month','Booking_month','Age'], axis=1, inplace=True)
```

```python
# df_assault_drop.to_csv("Resources/df_assault_drop.csv",index=True,header=True)
```

```python
df_assault_drop.head()
```

```python
# Create Columnsfor Days of the week
df_dummies = pd.get_dummies(df_assault_drop)
df_dummies.head()
```

```python
# df_dummies.to_csv("Resources/df_dummies.csv",index=True,header=True)
```

```python
df_combined = df_assault.copy()
```

```python
df_combined['day_of_week_Monday'] = df_dummies['day_of_week_Monday']
df_combined['day_of_week_Tuesday'] = df_dummies['day_of_week_Tuesday']
df_combined['day_of_week_Wednesday'] = df_dummies['day_of_week_Wednesday']
df_combined['day_of_week_Thursday'] = df_dummies['day_of_week_Thursday']
df_combined['day_of_week_Friday'] = df_dummies['day_of_week_Friday']
df_combined['day_of_week_Saturday'] = df_dummies['day_of_week_Saturday']
df_combined['day_of_week_Sunday'] = df_dummies['day_of_week_Sunday']
df_combined['CityRN_FOREIGN'] = df_dummies['CityRN_FOREIGN']
df_combined['CityRN_HOMELESS'] = df_dummies['CityRN_HOMELESS']
df_combined['CityRN_MIAMI'] = df_dummies['CityRN_MIAMI']
```
